# **Exploratory Data Analysis (EDA)**

#### **What is Exploratory Data Analysis?**
Exploratory Data Analysis (EDA) is a crucial step in the data analysis process that helps in understanding various aspects of the data before applying any machine learning models. It involves summarizing, visualizing, and identifying patterns, trends, or anomalies in the dataset.

#### **Objective of EDA**
The primary objective of EDA is:
- To **understand the data** structure, distribution, and key characteristics.
- To **identify faulty points**, such as missing values, outliers, and redundant information.
- To **clean the data**, ensuring there are no duplicates or inconsistencies.
- To **identify important variables** and relationships between them.
- To **remove unnecessary noise** that can hinder the accuracy of machine learning models.
- To **derive insights** and hypotheses that may guide further analysis.

### **Steps Involved in EDA**
1. **Understanding the Data**
   - Load the dataset and explore its structure.
   - Examine the first few and last few rows.
   - Get a summary of the dataset (mean, min, max, standard deviation, etc.).
   - Identify the number of rows and columns.
   - Check for unique values and categorical variables.

2. **Cleaning the Data**
   - Check for missing values.
   - Handle outliers (which may cause overfitting or underfitting).
   - Convert categorical variables if needed.
   - Drop irrelevant or redundant columns.

3. **Analyzing Relationships Between Variables**
   - Compute correlation matrices to understand dependencies.
   - Visualize distributions, trends, and patterns.
   - Generate pair plots, scatter plots, and box plots.
   - Detect outliers and examine their impact on the dataset.

EDA is an essential step to **detect patterns, clean data, and extract insights** before applying any machine learning models. A thorough EDA ensures that the dataset is ready for modeling, leading to **better accuracy and efficiency** in predictions.

### **What Are Outliers?**
Outliers are extreme values that significantly deviate from most data points. They can result from measurement errors, data entry mistakes, or genuine rare occurrences.

### **Characteristics of Outliers**  
- **Extreme Values** – Significantly higher or lower than most data points.  
- **High Influence** – Can distort statistical analyses and model predictions.  
- **Possible Causes** – Measurement errors, data entry mistakes, rare but valid occurrences.  
- **Detection Methods** – Boxplots, Z-scores, IQR method, and scatter plots.

### **Datasets That Typically Have Outliers**   
1. **Finance & Economics** – High variability due to market fluctuations and extreme wealth distribution.  
2. **Healthcare & Medicine** – Biological variations and rare medical conditions.  
3. **Social Sciences** – Diverse human behavior and socioeconomic disparities.  
4. **Engineering & Manufacturing** – Sensor malfunctions, defects, and system failures.  
5. **Sports & Performance Metrics** – Exceptional athlete performances and errors in measurements.  
6. **E-commerce & Retail** – Fraudulent transactions, bulk purchases, and pricing anomalies.  



##### **Check for Outliers**

````python
for column in studentacadprofile.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=studentacadprofile[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
````

## **Step 3: Analyzing Relationships Between Variables**

##### **Compute Correlation Matrix**

**Fixing the ValueError in Correlation Calculation**
**Error:**
```
ValueError: could not convert string to float: 'Kiana Lor'
```
**Reason:** The correlation function `corr()` expects numeric values, but some columns contain non-numeric data.

**Solution:** Select only numeric columns before computing correlation.
```python
correlation = studentacadprofile.select_dtypes(include=np.number).corr()
```

````python
# correlation = studentacadprofile.corr() --> Gives `ValueError`
correlation = studentacadprofile.select_dtypes(include=np.number).corr()
````

##### **Visualizing the Correlation Matrix with Heatmap**


````python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,
            annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
plt.title("Correlation Heatmap")
plt.show()
````

##### **Pair Plot for Visualizing Relationships**

````python
sns.pairplot(studentacadprofile)
plt.show()
````

##### **Scatter Plot to Explore Specific Relationships**

````python
sns.relplot(x='math.grade', y='language.grade', hue='gender', data=studentacadprofile)
plt.title("Math Grade vs Language Grade by Gender")
plt.show()
````

##### **Distribution Plot of Math Grades**

````python
sns.histplot(studentacadprofile['math.grade'], kde=True)
plt.title("Distribution of Math Grades")
plt.xlabel("Math Grade")
plt.show()
````

##### **Box Plot to Identify Outliers**

````python
sns.boxplot(x=studentacadprofile['math.grade'])
plt.title("Boxplot of Math Grades")
plt.show()
````
