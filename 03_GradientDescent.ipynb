# **Gradient Descent**
Gradient descent is an optimization algorithm commonly used to minimize a loss function in machine learning and deep learning models. It's an iterative method to find the minimum of a function (often the cost or loss function in machine learning), which can be used for a variety of purposes such as training models or adjusting the parameters of algorithms.


### 1. **Basic Idea**
The main idea behind gradient descent is to start with some initial parameter values and then update them iteratively to reduce the cost function value. The update rule is based on the negative gradient of the cost function with respect to the model parameters. In simple terms, we move in the direction that reduces the loss the fastest.

### 4. **Learning Rate (α)**
The learning rate controls how large the steps should be. A small learning rate leads to slow convergence, while a large learning rate might cause overshooting of the minimum or cause the algorithm to diverge.

To prevent this, various techniques have been introduced to adjust the learning rate dynamically, such as:

- **Learning Rate Decay**: Decreasing the learning rate as training progresses.
- **Momentum**: Adding a fraction of the previous update to the current update to smooth out the path and avoid oscillations.
- **Adaptive Methods**: Algorithms like Adam, RMSProp, and AdaGrad adjust the learning rate per parameter, making it more efficient for different features.

### 5. **Convergence and Local Minima**
Gradient descent can face challenges such as:

- **Local Minima**: If the function has multiple minima (points that are lower than the surrounding areas), gradient descent may get stuck in a local minimum instead of finding the global minimum.
- **Saddle Points**: Gradient descent can also be stuck at saddle points (points where the gradient is zero but are not minima).
  
To deal with these, techniques like **momentum**, **random restarts**, or more advanced methods like **Simulated Annealing** are often used to help escape local minima.

### 6. **Implementation in Pseudocode**

```text
Initialize θ randomly
Repeat until convergence:
    Compute gradient of J(θ)
    Update θ using the rule:
    θ := θ - α * gradient
```



### 7. **Advantages and Disadvantages**
#### **Advantages**:
- Simple and efficient for convex loss functions.
- Can be used for a wide variety of machine learning models.

#### **Disadvantages**:
- Requires choosing a good learning rate.
- Can be slow for large datasets or complex models.
- Can get stuck in local minima for non-convex loss functions.

Implementing the basic version of gradient descent for a simple cost function (like a linear regression problem) and gradually move toward more complex implementations like stochastic or mini-batch gradient descent.

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the cost function (e.g., a simple 2D quadratic function)
def cost_function(x, y):
    return x**2 + y**2  # A simple paraboloid

# Gradient of the cost function
def gradient(x, y):
    dfdx = 2*x  # Partial derivative with respect to x
    dfdy = 2*y  # Partial derivative with respect to y
    return np.array([dfdx, dfdy])

# Gradient Descent Algorithm
def gradient_descent(starting_point, learning_rate, num_iterations):
    points = [starting_point]
    point = starting_point
    for _ in range(num_iterations):
        grad = gradient(point[0], point[1])
        point = point - learning_rate * grad  # Update the point
        points.append(point)
    return np.array(points)

# Parameters
learning_rate = 0.1
num_iterations = 50
starting_point = np.array([2.0, 2.0])  # Starting point for gradient descent

# Run gradient descent
points = gradient_descent(starting_point, learning_rate, num_iterations)

# Create a grid for the cost function
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = cost_function(X, Y)

# Plot the surface and the path of gradient descent
fig = plt.figure(figsize=(14, 10))  # Increased figure size
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)

# Plot the path of gradient descent
ax.plot(points[:, 0], points[:, 1], cost_function(points[:, 0], points[:, 1]), color='r', marker='o')

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Cost Function')
ax.set_title('3D Gradient Descent Visualization')


plt.show()
````
